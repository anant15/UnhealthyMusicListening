```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Data_HUMS_Traits.csv")

data.columns = list(map(lambda it: it.strip("'"), data.columns))

data = data.drop(["BMRQ.Social_Reward", 
           "BMRQ.Musical_Seeking", 
           "BMRQ.Mood_Regulation", 
           "BMRQ.Sensory_Motor", 
           "BMRQ.Emotional_Evocation",
                 "HUMS.Healthy"], axis=1)
```

```python
X = data.drop('HUMS.Unhealthy', axis=1)
y = data['HUMS.Unhealthy']
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2, interaction_only=True, include_bias=False)

x_interaction = poly.fit_transform(X)
x_interaction_columns = poly.get_feature_names(input_features=X.columns)
interaction_df = pd.DataFrame(x_interaction, columns=x_interaction_columns)
```

```python
from statsmodels.regression import linear_model
import statsmodels.api as sm

interaction_df = sm.add_constant(interaction_df)
X = sm.add_constant(X)

model = linear_model.OLS(y, X).fit()
interaction_model = linear_model.OLS(y, interaction_df).fit()
```

```python
print(model.pvalues[model.pvalues < 0.05])
model.summary()
```

```python
print(interaction_model.pvalues[interaction_model.pvalues < 0.05])
interaction_model.summary()
```
